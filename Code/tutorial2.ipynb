```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
co2=pd.read_csv('../Data/co2-data/owid-co2-data.csv')
```

```python
co2.info()
```

```python
co2.describe()
```

```python
np.unique(co2['country'])
```

```python
us = co2[co2['country'] == 'United States']
```

```python
energy=us['energy_per_capita']*us['population']
yrus=us['year']
### kwhour in ExoJ
kwhour=1.e-15*3600
energy=energy*kwhour
plt.plot(yrus, energy)
plt.xlabel('Year')
plt.ylabel('Total Energy Use (EJ)')
country='United States'
plt.title(country)
```

```python
epcus=us['energy_per_capita']
plt.plot(yrus, epcus*1.e-6*3600) # GJoules
plt.xlabel('Year')
plt.ylabel('Energy Use Per Capita (GJ)')
plt.title(country)
```

```python
country='France'
df_country = co2[co2['country'] == country]
epcfr=df_country['energy_per_capita']
yrfr=df_country['year']
plt.plot(yrfr, epcfr*1.e-6*3600) # GJoules
plt.xlabel('Year')
plt.ylabel('Energy Use Per Capita (GJ)')
plt.title(country)
```

```python
country='India'
df_country = co2[co2['country'] == country]
epcn=df_country['energy_per_capita']
yrn=df_country['year']
plt.plot(yrn, epcn*1.e-6*3600) # GJoules
plt.xlabel('Year')
plt.ylabel('Energy Use Per Capita (GJ)')
plt.title(country)
```

```python
plt.plot(yrn, epcn*1.e-6*3600,label='India') # GJoules
plt.plot(yrus, epcus*1.e-6*3600,label='United States') # GJoules
plt.plot(yrfr, epcfr*1.e-6*3600,label='France') # GJoules
plt.xlabel('Year')
plt.ylabel('Energy Use Per Capita (GJ)')
plt.legend()
plt.yscale('log')
```

```python
country='High-income countries'
df_country = co2[co2['country'] == country]
epc=df_country['energy_per_capita']
yr=df_country['year']
plt.plot(yr, epc*1.e-6*3600,label='High-income')# GJoules
country='Low-income countries'
df_country = co2[co2['country'] == country]
epc=df_country['energy_per_capita']
yr=df_country['year']
plt.plot(yr, epc*1.e-6*3600,label='Low-income')# GJoules
country='Upper-middle-income countries'
df_country = co2[co2['country'] == country]
epc=df_country['energy_per_capita']
yr=df_country['year']
plt.plot(yr, epc*1.e-6*3600,label='Upper-middle-income')# GJoules
country='Lower-middle-income countries'
df_country = co2[co2['country'] == country]
epc=df_country['energy_per_capita']
yr=df_country['year']
plt.plot(yr, epc*1.e-6*3600,label='Lower-middle-income')# GJoules

plt.xlabel('Year')
plt.ylabel('Annual Energy Use Per Capita (GJ)')
plt.legend()
```

```python
country='High-income countries'
df_country = co2[co2['country'] == country]
epc=df_country['energy_per_capita']
pop=df_country['population']
yr=df_country['year']
plt.plot(yr, epc*pop*1.e-6*3600,label='High-income')# GJoules

country='Low-income countries'
df_country = co2[co2['country'] == country]
pop=df_country['population']
epc=df_country['energy_per_capita']
yr=df_country['year']
plt.plot(yr, epc*pop*1.e-6*3600,label='Low-income')# GJoules

country='Upper-middle-income countries'
df_country = co2[co2['country'] == country]
pop=df_country['population']
epc=df_country['energy_per_capita']
yr=df_country['year']
plt.plot(yr, epc*pop*1.e-6*3600,label='Upper-middle-income')# GJoules

country='Lower-middle-income countries'
df_country = co2[co2['country'] == country]
pop=df_country['population']
epc=df_country['energy_per_capita']
yr=df_country['year']
plt.plot(yr, epc*pop*1.e-6*3600,label='Lower-middle-income')# GJoules

plt.xlabel('Year')
plt.ylabel('Annual Energy Use (GJ)')
plt.legend()
```
