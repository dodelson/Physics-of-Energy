```python
h=6.63e-34
c=3.e8
kb=1.38e-23
hbar=h/(2*np.pi)
```

```python
logwavel=np.arange(-8.,-4.,.1)
wavel=10.**logwavel
prefactor=hbar*(2*np.pi*c)**2/wavel**5
T=279.
x=h*c/(wavel*kb*T)
fearth=prefactor/(np.exp(x)-1.)
plt.plot(wavel,fearth,label='T=279K')
T=5800.
x=h*c/(wavel*kb*T)
fsun=prefactor/(np.exp(x)-1.)
plt.plot(wavel,fsun,label='T=5800K')
plt.axis([1.e-8,1.e-4,1.,1.e15])
plt.xscale('log')
plt.yscale('log')
plt.ylabel('dE/dtdA$d\lambda (W/m^2$/m)')
plt.xlabel('$\lambda (m)$')
plt.legend()
plt.savefig('../bb.png')
```

Multiply by the surface area of Earth to get the total outgoing radiation. To get the total solar flux, multiply by the surface area of the Sun to get the total luminosity but then we get only the area of the Earth divided by $4\pi R^2$, the imaginary sphere that contains us. So,


```python
Rearth=6.38e6
Rsun=696.e6
Rorbit=147.97e9

T=279.
x=h*c/(wavel*kb*T)
fearth=prefactor/(np.exp(x)-1.)
T=5800.
x=h*c/(wavel*kb*T)
fsun=prefactor/(np.exp(x)-1.)

fearth*=4*np.pi*Rearth**2
fsun*=Rsun**2*Rearth**2/Rorbit**2
plt.plot(wavel,fsun,label='Incoming Solar Flux')
plt.plot(wavel,fearth,label='Outgoing Radiation')
plt.axis([1.e-7,1.e-4,1.e20,1.e23])
plt.xscale('log')
plt.yscale('log')
plt.ylabel('dE/dt$d\lambda$ (W/m)')
plt.xlabel('$\lambda (m)$')
plt.legend()
plt.savefig('../balance.png')
```

```python
Rearth=6.38e6
Rsun=696.e6
Rorbit=147.97e9

T=279.
x=h*c/(wavel*kb*T)
fearth=prefactor/(np.exp(x)-1.)
T=5800.
x=h*c/(wavel*kb*T)
fsun=prefactor/(np.exp(x)-1.)

#fearth*=4*np.pi*Rearth**2
fsun*=Rsun**2/(4*Rorbit**2)
plt.plot(wavel*1e9,fsun/1.e9,label='Incoming Solar Flux')
plt.plot(wavel*1e9,fearth/1.e9,label='Outgoing Radiation')
plt.axis([100,1e5,1.e-12,1.])
plt.xscale('log')
plt.yscale('log')
plt.ylabel('dP/dA$d\lambda$ (W/m$^2$/nm)')
plt.xlabel('$\lambda (nm)$')
plt.legend()
plt.savefig('../balance.png')
```
