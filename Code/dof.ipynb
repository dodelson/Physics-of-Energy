```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
x=np.arange(0.01,10.,.1)
E0=2*3.14*2.e-5/15.e-4 #eV
#wavel=2*3.14*2.e-5/E0 # lambda=c/nu = hc/E =2pi hbar c/E = 2pi *2.e-5 eV-cm/E
E = E0*(1/(np.exp(x)-1.))
print(E0)
wavel=4000 # E/hc in cm^{-1}
en=4000.*(2*3.14*2.e-5)
print(en)
```

```python
T=E0/x
plt.plot(T,E+1.5*T,label='Internal Energy')
plt.plot(T,T*2.5,label='Two Degrees of Freedom')
plt.yscale('log')
plt.legend()
plt.xscale('log')
plt.axis([.01,5,.01,5])
plt.xlabel('$k_bT$ (eV)')
plt.ylabel('E-E_0 (eV)')
```

```python
T=E0/x
plt.plot(T*11605,E+1.5*T,label='Internal Energy')
plt.plot(T*11605,T*2.5,label='Two Degrees of Freedom')
plt.yscale('log')
plt.legend()
plt.xscale('log')
#plt.axis([.1,5,.1,5])
plt.xlabel('$T$ (K)')
plt.ylabel('E-E_0 (eV)')
```

```python
#rotational
Er=7.69e-24/1.6e-19
print(Er)
Erot=T#/Er
plt.plot(T*11605,E+Erot+1.5*T,label='Total Energy Per Particle')
plt.plot(T*11605,T*2.5,label='Rotational Degrees of Freedom')
plt.plot(T*11605,T*3.5,label='Rot + Vib Degrees of Freedom')
plt.yscale('log')
plt.legend()
plt.xscale('log')
plt.axis([200.,5000,.04,4])
plt.xlabel('$T$ (K)')
plt.ylabel('E-E_0 (eV)')
plt.savefig('../Figures/dof.png')
```
