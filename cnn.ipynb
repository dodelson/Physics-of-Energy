```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
filename='Data/co2_mm_mlo.txt'
day=np.loadtxt(filename).T[2]
co2=np.loadtxt(filename).T[3]
gtc=2.13 # global tons of carbon for 1 ppm
plt.plot(day,co2*gtc)
plt.xlabel('Year')
plt.ylabel('$CO_2$ (GigaTons Carbon)')
%pwd
```

There is more CO2 in the atmosphere in Northern Spring and less in the Fall. This reflects the fact that throughout the winter the plants are decaying and returing CO2 to the atmosphere. Then in the Spring, the plants start recapturing the CO2 to grow so that by Fall the atmospheric CO2 is a minimum.

```python
def rmnan(db):
    dbnew=pd.DataFrame()
    emissions=db['co2']
    gdp=db['gdp']
    population=db['population']
    n=np.size(emissions)
    print(n)
    dbnew=db.dropna(axis=0,subset=['co2','gdp','population'])
    emissions=dbnew['co2']
    print(n,np.size(emissions))
    return dbnew
```

```python
world=co2[co2['country']=='World']
year=world['year']
emit=world['co2']
plt.plot(year,emit/3.664/1.e3) # divide by 1.e3 because it is millions not billons
plt.xlabel('Year')
plt.ylabel('Global Carbon Emissions (GTons)')
```

```python
n=np.size(emit)
cum=np.zeros(n)
for i in range(n):
    cum[i]=np.sum(emit[:i])
plt.plot(year,cum/3.664/1.e3)
plt.xlabel('year')
plt.ylabel('Cumulative Anthropogenic Carbon Emissions (Gtons)')
```

```python
co22019=co2[co2['year']==2010]
co22019=co22019[co22019['country']!='World']
emissions=co22019['co2']
gdp=co22019['gdp']
population=co22019['population']
#Regression
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
## clean up nans
n=np.size(emissions)
print(n)
for i in gdp.index:
    if (np.isnan(gdp[i]) or np.isnan(emissions[i]) or np.isnan(population[i])): 
        del gdp[i]
        del emissions[i]
        del population[i]
n=np.size(emissions)
print(n)
emissions=emissions.values/1.e3/3.664
gdp=gdp.values/1.e12
population=population.values/1.e9
#gdp=gdp.reshape(-1,1)
#print(emissions)
#emissions=emissions.reshape(-1,1)
X=np.transpose([np.log(gdp),np.log(population)])
print(np.shape(X),np.shape(emissions))
lin_reg.fit(X,np.log(emissions))
print(lin_reg.coef_,lin_reg.intercept_)
x=np.arange(0,20,.0001)
#x.reshape(-1,1)
plt.scatter(np.exp(lin_reg.predict(X)),(emissions))#,label='2D Regression Fit')
#plt.scatter(gdp,emissions,label='Data')
plt.xlabel('Predicted Emissions')
plt.ylabel('Actual Carbon Emissions (GTons)')
plt.plot([5.e-5,3],[5.e-5,3])
#plt.axis([5.e-4,3,1.e-4,3])
plt.yscale('log')
plt.xscale('log')
print(lin_reg.score(X,np.log(emissions)))
```

Just do a "fully connected network". That is, for each country, take its emissions over all years as features and then see how well a simple regression analysis does. The simplest thing would be to just predict that the coming year's emissions were equal to the previous years. This way, you get to weight all previous years. Still not taking into account GDP or population, and doing a simple linear regression in these "features" (the emissions in previous years).

```python
# Find out the sizes of the different countries' data sets.
country=co2['country'].values
country=np.unique(country)
nyears=[]
for c in country:
    tmp=co2[co2['country']==c]
    y=tmp['year'].values
    nyears.append(np.size(y))
plt.hist(nyears)
```

So one problem is that all the countries have a different number of years, in this case a different number of features. For simplicity let's just use the past 50 years. 

```python
## For the features, use something that has similar values: Emissions/(gdp^1.25/pop^.25)
X=[]
for c in country:
    tmp=co2[co2['country']==c]
    y=tmp['year'].values
    if np.size(y) >= 50:
        c50=tmp['co2'][-50:]
        g=tmp['gdp'][-50:]
        p=tmp['population'][-50:]
        X.append(c50*p**.25/g**1.25/1.e-10)
X=np.array(X)
```

```python
from sklearn.preprocessing import StandardScaler as SS
#print(np.shape(X))
#scaler=SS()
#scaler.fit(X)
Xsc=X#scaler.transform(X)
#yy=scaler.get_params()
#print(yy)
```

```python
X_train,y_train=Xsc[:100,:49],Xsc[:100,49:]
X_test,y_test=Xsc[100:,:49],Xsc[100:,49:]
print(np.shape(X_train))
print(X_train[0,0])
```

The baseline error here is just to take the previous value of the data. Figure out what the loss function would be in that case

```python
mse=0.
for i in range(100):
    mse+=(X_train[i,-1]-y_train[i])**2
print(mse/100)
```

```python
## This is akin to a linear regression model
import tensorflow as tf
import tensorflow 
from tensorflow import keras
model1=keras.models.Sequential([
    keras.layers.Flatten(input_shape=[49,1]),
#    keras.layers.SimpleRNN(20,return_sequences=True),
#    keras.layers.SimpleRNN(20),
    keras.layers.Dense(1)#,activation='Softmax')
])
model1.summary()
```

```python
model1.compile(loss='mse')#,optimizer='sgd')
history=model1.fit(X_train,y_train,epochs=500)
print(model1.evaluate(X_train,y_train))
pd.DataFrame(history.history).plot()
plt.yscale('log')
```

```python
#So, after 500 epochs, it's still above .001 .. and choppy. Try for another 1000
history=model1.fit(X_train,y_train,epochs=1000)
print(model1.evaluate(X_train,y_train))
pd.DataFrame(history.history).plot()
plt.yscale('log')
```

Does dip below but too choppy, so try a more complex neural network, akin to the one on page 505.

```python
Xt=np.expand_dims(X_train, axis=2)
print(np.shape(Xt))
model2=keras.models.Sequential([
    keras.layers.SimpleRNN(1,input_shape=[None,1])
])
```

```python
model2.compile(loss='mse')#,optimizer='sgd')
history=model2.fit(Xt,y_train,epochs=500)
print(model2.evaluate(Xt,y_train))
pd.DataFrame(history.history).plot()
plt.yscale('log')
```

That does indeed beat the simple thing with no choppiness

```python
i=2
#print(y_train[0],X_train[0,0])
y_predict=model2.predict(Xt)
plt.plot(range(49),X_train[i,:])
plt.scatter(range(49,50),y_train[i,:],label='True')
plt.scatter(range(49,50),y_predict[i,:],label='RNN')
plt.legend()
print(keras.losses.mean_squared_error(y_predict[i],y_train[i]))
print(keras.losses.mean_squared_error(X_train[i,-1],y_train[i]))

print(y_predict[i,:],y_train[i,:],X_train[i,-1])

```

The above (i=2) is an interesting example because the naive prediction is too low; the model seems to understand that the previous point was a random downward fluctuation and the overall trend is positive. In this case, the model does better than naive.
